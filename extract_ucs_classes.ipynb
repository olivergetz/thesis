# Extract UCS Classes

This notebook extracts UCS classes from the official "UCS Full" spreadsheet for use during classification as conditioning text. These are saved to .json objects for easy conversion between class index and class name.

```python
import pandas as pd
import json

df = pd.read_csv("./experiments/data/ucs_official/ucs_full_official.csv")
```

```python
int_to_class = {}
class_to_int = {}
for i, class_name in enumerate(df['Category'].unique().tolist()):
    class_to_int[class_name.lower()] = i
    int_to_class[i] = class_name.lower()

ucs_class_conversion = {}
ucs_class_conversion['class_to_int'] = class_to_int
ucs_class_conversion['int_to_class'] = int_to_class

with open('data/ucs_classes.json', 'w', encoding='utf-8') as f:
    json.dump(ucs_class_conversion, f, ensure_ascii=False, indent=4)
```
