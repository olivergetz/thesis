# Explore Clap Models

This notebook is used to check the size of the clap objects and log memory constraints. Stop all kernels before running this notebook.

```python
import torch
from msclap import CLAP
import laion_clap
from experiments.utils.set_torch_device import set_torch_device
```

```python
device, device_name = set_torch_device()
torch.cuda.empty_cache()

start = torch.cuda.memory_allocated(device)
laion_clap_model = laion_clap.CLAP_Module(enable_fusion=False)
laion_clap_model.load_ckpt() # Load default checkpoint: 630k-audioset-best.pt
laion_model_mem = start - torch.cuda.memory_allocated(device)

start = torch.cuda.memory_allocated(device)
msclap_model = CLAP(version = '2023', use_cuda=True) # version can be 2022 or 2023
msclap_model_mem = start - torch.cuda.memory_allocated(device)

print(f"Laion CLAP size in gigabytes for {device_name}: {laion_model_mem*1e-9}")
print(f"MSCLAP size in gigabytes for {device_name}: {msclap_model_mem*1e-9}")
```
